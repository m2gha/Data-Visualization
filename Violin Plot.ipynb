```python
#Violin PLOT

import matplotlib.pyplot as plt

import pandas as pd



wine_quality = pd.read_csv('winequality.csv',delimiter=';')

data = [wine_quality['alcohol'], wine_quality['fixed acidity'],wine_quality['quality']]

plt.violinplot(data, showmeans=True)

plt.show()
```

```python
import matplotlib.pyplot as plt

import pandas as pd



alcohol = [1,2,3,4,5,6,7,1010]
fixed_acidity=[5,6,89,94,2,5,6,8]
quality=[19,20,17,21,45,32,12,45]


#data = [alcohol,fixed_acidity,quality]

fig,(ax1,ax2,ax3)=plt.subplots(nrows=1,ncols=3)

ax1.violinplot(alcohol,showmedians=True)
ax1.set_title('Alcohol')

ax2.violinplot(fixed_acidity,showmedians=True)
ax2.set_title('Fixed Acidity')

ax3.violinplot(quality,showmedians=True,showmeans=True)
ax3.set_title('Quality')


plt.show()
```
